### 3.1.3 Texts

I will analyse the attribute objectives of the courses to get a better understanding about the courses. At first I needed to transfor the list into  one text, and then investigate the frequencies of each word.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
from collections import defaultdict
from operator import itemgetter
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
```

```python
#create a dataframe for the stemming
from nltk.tokenize import RegexpTokenizer
from nltk.stem import SnowballStemmer
#tokenizer=RegexpTokenizer(r'[a-zA-Z]+')
from nltk.tokenize import word_tokenize
StopWords=set(stopwords.words('english')+list(punctuation)+['’','’s',"'s"])
stemmer=SnowballStemmer('english')

def tokenize(text):
    #return [stemmer.stem(word) for word in tokenizer.tokenize(text.lower())]
    return [stemmer.stem(word) for word in word_tokenize(text.lower())]

def tokenize_only(text):
    #return [word for word in tokenizer.tokenize(text.lower())]
    return [word for word in word_tokenize(text.lower())]
def combine_list(l):
    new_str=""
    for item in l:
        new_str=new_str+' '+str(item) 
    return new_str

objectives_text=df_courses['objectives'].apply(combine_list)
total_stemmed = []
total_tokenized = []
for i in objectives_text:
    obj_stemmed = tokenize(i) 
    total_stemmed.extend(obj_stemmed) 
    
    obj_tokenized = tokenize_only(i)
    total_tokenized.extend(obj_tokenized)
    
vocab_frame_orig = pd.DataFrame({'words': total_tokenized}, index = total_stemmed)
print ('there are ' + str(vocab_frame_orig.shape[0]) + ' items in vocab_frame')
vocab_frame_orig.head()    
```

```python
vocab_frame=vocab_frame_orig.copy()
vocab_frame=vocab_frame.reset_index()
vocab_frame.columns = ['index','words']
vocab_frame=vocab_frame.drop_duplicates(subset='index', keep='first').set_index('index')
vocab_frame.shape
```

```python
#check the most common words by each subcategory
def get_words(df):
    words_dict=defaultdict(int)
    for items in df['objectives'].values:
        for item in items:
            words=tokenize(item)
            for word in words:
                if word not in StopWords:
                    try:
                        word_new=vocab_frame.loc[word].values.tolist()[0]
                    except:
                        word_new==word
                    words_dict[word_new] += 1
    return sorted(words_dict.items(), key=itemgetter(1), reverse=True)
```

```python
for subcat in df_courses['primary_subcategory'].unique():
    temp=df_courses[df_courses['primary_subcategory']==subcat]
    top_words=get_words(temp)[:25]
    plt.subplots(figsize=(10,8))
    plt.barh(range(len(top_words)), [val[1] for val in top_words], align='center')
    plt.yticks(range(len(top_words)), [val[0] for val in top_words])
    plt.title(subcat)
    plt.xlabel("Number of occurences")
```

```python
from wordcloud import WordCloud

for subcat in df_courses['primary_subcategory'].unique():
    temp=df_courses[df_courses['primary_subcategory']==subcat]
    top_words=dict(get_words(temp)[:25])
    plt.subplots(figsize=(10,8))
    wordcloud = WordCloud(width=900,height=500, margin=0).generate_from_frequencies(top_words)

    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(subcat)
```
